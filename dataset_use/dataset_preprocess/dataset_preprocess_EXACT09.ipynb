resort the image names of some cases排序

```python
def resort_names(case_dict, chosen_name, indicator):
    new_path_idx_list = []
    for path in case_dict[chosen_name]:
        name = path.split("/")[-1]
        new_path_idx_list.append(int(name.split(indicator)[1]))
    new_path_idx_list = np.array(new_path_idx_list)
    locs = np.argsort(new_path_idx_list)

    new_path_list = []
    for i in locs:
        new_path_list.append(case_dict[chosen_name][i])

    case_dict[chosen_name]=new_path_list
```

```python
def loadFile(filename):
    ds = sitk.ReadImage(filename)
    img_array = sitk.GetArrayFromImage(ds)
    frame_num, width, height = img_array.shape
    return img_array, frame_num, width, height
#loadFile(filename) 接受文件名作为输入，使用 SimpleITK 库读取图像，将其转换为 numpy 数组，并返回该数组以及其维度。
def get_3d_img_for_one_case(img_path_list):
    img_3d=[]
    for idx, img_path in enumerate(img_path_list):
        print("progress: "+str(idx/len(img_path_list))+"; "+str(img_path), end="\r")
        img_slice, frame_num, _, _ = loadFile(img_path)
        assert frame_num==1
        img_3d.append(img_slice)
    img_3d=np.array(img_3d)
    return img_3d.reshape(img_3d.shape[0], img_3d.shape[2], img_3d.shape[3])
#get_3d_img_for_one_case(img_path_list) 接受图像文件路径列表作为输入，
# 使用 loadFile() 加载每个图像，并将其附加到 3D numpy 数组 img_3d。
# 然后，它将数组重新调整为维度 (图像数量，高度，宽度) 并返回它。
```

```python
# preprocess and save one case
img_3d = get_3d_img_for_one_case(train_case_dict["CASE20"])
sitk.WriteImage(sitk.GetImageFromArray(img_3d), output_file_path+"/train/CASE20.nii.gz")
```

```python
# get labels of the training data
label_file_path = "/mnt/share102/cs22-wangc/EXACT09/AirwayAnnotation/airways_EXACTtrain"
# .mhd文件是一种用于存储医学图像数据的文件格式。
# 该格式通常与.mha文件配对使用，后者存储实际的图像数据。
# .mhd文件包含元数据，如图像的维度、像素间距、像素类型、坐标系和数据偏移等。这些元数据描述了图像数据的属性和存储方式。

# .mhd文件是MetaImage文件格式的一部分，
# 由ITK（Insight Segmentation and Registration Toolkit）开发，支持跨平台和跨语言使用。
# 因此，.mhd文件在医学图像处理和分析中被广泛使用，例如计算机断层扫描（CT）和磁共振成像（MRI）等领域。
# 这段代码中，首先定义了一个变量label_file_path，表示标签数据所在的文件路径。
# 然后，对于train_case_dict字典中的每一个key，依次读取该key对应的.mhd文件，并将读取到的数据保存为一个.nii.gz文件。

# 具体来说，loadFile()函数是一个用于读取.mhd文件的函数，
# 它的输入参数是.mhd文件的完整路径，返回值包括读取到的数据以及其他一些元信息。
# 在这里，只需要用到了读取到的数据，保存在变量img_label_arr中。
# 然后，使用sitk.WriteImage()函数将img_label_arr保存为.nii.gz文件，
# 输出路径为output_file_path+"/train_label/"+case+'_label.nii.gz'，
# 其中case为当前循环的train_case_dict的key。最后，循环结束后就可以得到所有.mhd文件对应的.nii.gz文件。
for case in train_case_dict.keys():
    print(case)
    img_label_arr, _, _, _ = loadFile(label_file_path+'/'+case+'.mhd')
    sitk.WriteImage(sitk.GetImageFromArray(img_label_arr),
                    output_file_path+"/train_label/"+case+'_label.nii.gz')
#在医学图像领域中，常见的图像数据格式包括.mhd和.nii.gz等。其中，.mhd格式通常是一个包含图像元数据和二进制图像数据的文件，而.nii.gz格式则是一个压缩的二进制图像数据文件。这两种格式都可以存储医学图像数据，但具体使用哪种格式通常取决于应用场景和需求。
# 在代码中，使用SimpleITK库中的WriteImage函数可以将一个SimpleITK图像对象写入到磁盘上的一个文件中。对于.nii.gz格式的文件，SimpleITK库内置了对该格式的支持，可以直接将一个SimpleITK图像对象保存为.nii.gz文件。而对于.mhd格式的文件，SimpleITK库同样也提供了读取和写入该格式文件的函数。
```
