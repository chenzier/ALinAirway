### Get pre-cropped EXACT09 dataset info

```python
Precrop_dataset_for_train_path = "/mnt/share102/cs22-wangc/Precrop_dataset_for_EXACT09"
Precrop_dataset_for_train_raw_path = Precrop_dataset_for_train_path+"/image"
Precrop_dataset_for_train_label_path = Precrop_dataset_for_train_path+"/label"

raw_case_name_list = os.listdir(Precrop_dataset_for_train_raw_path)
label_case_name_list = os.listdir(Precrop_dataset_for_train_label_path)

assert raw_case_name_list == label_case_name_list

data_dict = dict()

for idx, name in enumerate(raw_case_name_list):
    print("process "+str(name)+" | "+str(idx/len(raw_case_name_list)), end="\r")
    data_dict[name.split(".")[0]]={}
    data_dict[name.split(".")[0]]["image"]=Precrop_dataset_for_train_raw_path+"/"+name
    data_dict[name.split(".")[0]]["label"]=Precrop_dataset_for_train_label_path+"/"+name

    # 这段代码对支气管在CT图像中的分割结果进行了处理，计算了三个指标：总像素数、边缘像素数和内部像素数，
    # 并将这些指标存储在一个字典对象data_dict中
    label_temp = np.load(Precrop_dataset_for_train_label_path+"/"+name)
    # 使用 np.load() 函数加载标签数据是因为这里的标签数据是存储在.npy格式的文件中的。
    # .npy是numpy中的一种文件格式，用于存储numpy数组的二进制文件格式，
    # 可保存numpy数组的shape，dtype等信息，读取和写入速度快，并且可以在不损失精度的情况下压缩数据，因此非常适合存储和读取大型数组数据。
    data_dict[name.split(".")[0]]["airway_pixel_num"]=np.sum(label_temp)
    #计算label_temp中值为1的像素的数量，并将其作为data_dict字典中对应元素的"airway_pixel_num"键的值。
    
    label_temp=np.array(label_temp, dtype=np.uint32, order='F')
    #将label_temp转换为数据类型为np.uint32（32位无符号整数类型）的数组，并按Fortran的内存布局（即列优先）进行存储。
    # 在这里将数组按照Fortran的内存布局进行存储，即按列优先顺序存储，是因为NumPy中的默认存储方式是按行优先顺序存储，也称为C语言的内存布局。
    # 但是在一些科学计算中，特别是在图像处理等领域，按列优先的Fortran的内存布局可以更好地利用硬件的内存缓存，从而提高代码的运行效率。
    # 此外，在一些数值计算库和科学计算软件中，也常常采用Fortran风格的内存布局。
    
    # 使用edt.edt函数计算标签中每个像素到最近边界像素的欧几里得距离，并将结果存储到label_temp_edt中。
    # 第4步的代码主要是利用欧几里得距离变换 (Euclidean Distance Transform, EDT) 对图像的二值化掩膜进行处理，
    # 并计算出空气道的边界像素数和内部像素数。下面对代码中的每一部分进行详细说明：
    # 首先，代码通过调用edt函数对标签数据进行处理。
    # edt函数是edt模块中的函数，可以对二值图像进行欧几里得距离变换。
    # 在这里，函数的输入是label_temp变量，即原始的标签数据。
    # 参数black_border指定是否将边界像素的距离设为0。
    # 参数order指定数组的存储方式，即按Fortran的内存布局进行存储。
    # 参数parallel指定是否启用并行计算。函数的返回值是经过欧几里得距离变换后的图像数据。
    label_temp_edt=edt.edt(
        label_temp,
        black_border=True, order='F',
        parallel=1)
    # label_temp_edt是通过计算距离变换得到的结果，其元素不再是二值标签图像，
    # 而是对于每个像素点，表示该点到最近背景像素点的距离。
    # 因此，距离为1的像素点表示的是位于二值标签边界上的像素点，
    # 距离大于1的像素点表示的是位于二值标签内部的像素点。
    # 等于0的点表示不属于肺部内部的区域。
    # 因此，在计算airway_pixel_num_inner时需要选取距离大于1的像素点。
    
    
    #  接下来，代码分别计算了airway的边界像素数和内部像素数。
    # 对于边界像素数，代码使用了np.where函数来查找EDT图像中像素值为1的位置，并使用len函数计算出该位置数量。
    # 对于内部像素数，代码使用了np.where函数来查找EDT图像中像素值大于1的位置，并使用len函数计算出该位置数量。
    # 最后，将计算出的空气道的边界像素数和内部像素数分别保存到data_dict字典中的相应位置。
    data_dict[name.split(".")[0]]["airway_pixel_num_boundary"] = len(np.where(label_temp_edt==1)[0])
    data_dict[name.split(".")[0]]["airway_pixel_num_inner"] = len(np.where(label_temp_edt>1)[0])




save_obj(data_dict, "abc_dataset_info_EXACT09_crops_128")#如果文件名不存在，会在NaviAirway文件夹下自动创建
```

### 从dataset_info获取training strategy 首先，在high generations上train，然后在low generations上tranin 然后重复这个步骤
### Get dataset_info for iterative training strategy. First, train with higher frequency on airways of high generations. Next, train with higher frequency on airways of low generations. And repeat the process.

```python
#这一块代码看不太懂
data_dict_org=train_test_set_dict_EXACT09_LIDC_IDRI_128['train']

import copy
data_dict_extended = copy.deepcopy(data_dict_org)

is_more_big = True # higher freq on airways of low gen (thicker airways)
copy_times_I = 10
#从这步可以看出airway_pixel_num、airway_pixel_num_inner是用于分辨大小气道的
for idx, case in enumerate(data_dict_org.keys()):
    if data_dict_org[case]["airway_pixel_num"]>0:
        if is_more_big:
            copy_times_II = np.ceil(data_dict_org[case]["airway_pixel_num_inner"]/data_dict_org[case]["airway_pixel_num_boundary"])
        else:
            if data_dict_org[case]["airway_pixel_num_inner"]==0:
                copy_times_II = np.ceil(data_dict_org[case]["airway_pixel_num_boundary"])
            else:
                copy_times_II = np.ceil(data_dict_org[case]["airway_pixel_num_boundary"]/data_dict_org[case]["airway_pixel_num_inner"])

        for i in range(int(copy_times_I*copy_times_II)):
            data_dict_extended[case+"_copy_"+str(i+1)]=data_dict_org[case]
if is_more_big:
    save_obj(data_dict_extended, "abc_train_dataset_info_EXACT09_LIDC_IDRI_crops_128_extended_"+"more_low_gen_"+str(copy_times_I))
else:
    save_obj(data_dict_extended, "abc_train_dataset_info_EXACT09_LIDC_IDRI_crops_128_extended_"+"more_high_gen_"+str(copy_times_I))
```
