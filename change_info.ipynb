```python
#执行这个文件必须在python 3.8版本下
# To-do：None
import os
import numpy as np
import pickle
import edt

def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
#这里提供了一个函数去修改pkl文件
#如将地址 '/mnt/wangc/LIDC/Precrop_dataset_for_LIDC-IDRI/image/LIDC_IDRI_0529_46.nii.gz'修改为'test123/image/LIDC_IDRI_0529_46.nii.gz‘
def change_obj(old_name,new_name,addr):
    """
    变量说明:
    old_name:原先的pkl文件地址名称
    addr:需要修改的地址
    new_name:生成的新字典保存的位置（保存为一个pkl文件）
    
    """
    obj = load_obj(old_name)
    new_dict = {}
    for key, values in obj.items():
        new_values = {}
        for subkey, path in values.items():
            if 'image' in subkey or 'label' in subkey:
                new_values[subkey] = addr +str(subkey)+path.split(subkey)[1]
            else:
                new_values[subkey] = path
        new_dict[key] = new_values
    save_obj(new_dict, new_name)
```
